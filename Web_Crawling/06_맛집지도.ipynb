# 식신에서 찾은 맛집 -> 지도위에 표시하기

```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('맛집2.tsv', sep='\t')
df.head()
```

## 배경지도 그리기

```python
map = folium.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=16)
map
```

```python
#Marker
map = folium.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=16)
for i in df.index:
    folium.Marker(
        location=[df.lat[i], df.lng[i]],
        popup=df['상호'][i],
        tooltip=df['메뉴'][i]
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>산격동 맛집 - 맛집추천</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
# Circle 
map = folium.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=16)
for i in df.index:
    folium.Circle(
        radius=20,
        location=[df.lat[i], df.lng[i]],
        popup=df['상호'][i],
        tooltip=df['메뉴'][i],
        color='blue',
        fill=True,
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>산격동 맛집 - 맛집추천</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
# Circle 
map = folium.Map(location=[df.lat.mean(),df.lng.mean()],zoom_start=16, tiles= "Stamen Toner")# tiles= "Stamen Toner" 흑백 모드
for i in df.index:
    folium.Circle(
        radius=20,
        location=[df.lat[i], df.lng[i]],
        popup=df['상호'][i],
        tooltip=df['메뉴'][i],
        color='blue',
        fill=True,
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>산격동 맛집 - 맛집추천</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
